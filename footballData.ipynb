Get Team Positions after all matches have been played.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('premier_league22_23.csv')

# Calculate goal difference for each match
df['GD'] = df['FTHG'] - df['FTAG']

# Initialize 'Home_Points' and 'Away_Points' columns with zeros
df['Home_Points'] = 0
df['Away_Points'] = 0

# Update points and allocate to home and away teams based on match result
def update_points(row):
    if row['GD'] > 0:  # Home team wins
        row['Home_Points'] = 3
    elif row['GD'] < 0:  # Away team wins
        row['Away_Points'] = 3
    else:  # Draw
        row['Home_Points'] = 1
        row['Away_Points'] = 1
    return row

# Apply the function to each row to calculate points and update home and away team points
df = df.apply(update_points, axis=1)

# Group by home team and sum their total points and calculate overall goal difference
Standings_home = df.groupby('HomeTeam').agg({
    'Home_Points': 'sum',
    'GD':'sum'
}).reset_index()
Standings_home.rename(columns={'HomeTeam': 'Team'}, inplace=True)

# Group by away team and sum their total points and calculate overall goal difference
Standings_away = df.groupby('AwayTeam').agg({
    'Away_Points': 'sum',
    'GD':'sum'
}).reset_index()
Standings_away.rename(columns={'AwayTeam': 'Team'}, inplace=True)

# Merge the dataframes to get overall stats
Standings = pd.merge(Standings_home, Standings_away, on='Team', how='outer')

# Fill NaN values with 0, as they indicate that a team didn't play any home or away matches
Standings.fillna(0, inplace=True)

# Calculate overall total points and goal difference
Standings['Total_Points'] = Standings['Home_Points'] + Standings['Away_Points']
Standings['Overall_Goal_Difference'] = Standings['GD_x'] + Standings['GD_y']

# Drop unnecessary columns
Standings.drop(columns=['GD_x', 'GD_y'], inplace=True)

# Sort teams by total points and overall goal difference
sorted_teams = Standings.sort_values(by=['Total_Points', 'Overall_Goal_Difference'], ascending=[False, False])

# Reset the index for better readability
sorted_teams.reset_index(drop=True, inplace=True)

# Save the sorted DataFrame to a new CSV file
sorted_teams.to_csv('pl_standings_22-23.csv', index=False)

# Optionally, display the sorted DataFrame
print(sorted_teams)
```

Getting the overall standings of the 20 teams

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('premier_league22_23.csv')

# Calculate goal difference for each match
df['GD'] = df['FTHG'] - df['FTAG']

# Initialize columns for wins, losses, draws, and total matches played
df['Home_Won'] = 0
df['Home_Drawn'] = 0
df['Home_Lost'] = 0
df['Away_Won'] = 0
df['Away_Drawn'] = 0
df['Away_Lost'] = 0

# Update results for home and away teams
def update_results(row):
    if row['GD'] > 0:  # Home team wins
        row['Home_Won'] = 1
        row['Away_Lost'] = 1
    elif row['GD'] < 0:  # Away team wins
        row['Home_Lost'] = 1
        row['Away_Won'] = 1
    else:  # Draw
        row['Home_Drawn'] = 1
        row['Away_Drawn'] = 1
    return row

# Apply the function to each row to update results
df = df.apply(update_results, axis=1)

# Group by home team and calculate statistics
home_stats = df.groupby('HomeTeam').agg({
    'Home_Won': 'sum',
    'Home_Drawn': 'sum',
    'Home_Lost': 'sum',
    'FTHG': 'sum',  # Total goals scored at home
    'FTAG': 'sum'   # Total goals conceded at home
}).reset_index()
home_stats.rename(columns={'HomeTeam': 'Team', 'FTHG': 'Goals_Scored_Home', 'FTAG': 'Goals_Conceded_Home'}, inplace=True)

# Group by away team and calculate statistics
away_stats = df.groupby('AwayTeam').agg({
    'Away_Won': 'sum',
    'Away_Drawn': 'sum',
    'Away_Lost': 'sum',
    'FTAG': 'sum',  # Total goals scored away
    'FTHG': 'sum'   # Total goals conceded away
}).reset_index()
away_stats.rename(columns={'AwayTeam': 'Team', 'FTAG': 'Goals_Scored_Away', 'FTHG': 'Goals_Conceded_Away'}, inplace=True)

# Merge home and away statistics
team_stats = pd.merge(home_stats, away_stats, on='Team', how='outer')

# Fill NaN values with 0
team_stats.fillna(0, inplace=True)

# Calculate total matches played, total won, total drawn, total lost, and goal difference
team_stats['Total_Matches_Played'] = team_stats['Home_Won'] + team_stats['Home_Drawn'] + team_stats['Home_Lost'] + team_stats['Away_Won'] + team_stats['Away_Drawn'] + team_stats['Away_Lost']
team_stats['Total_Won'] = team_stats['Home_Won'] + team_stats['Away_Won']
team_stats['Total_Drawn'] = team_stats['Home_Drawn'] + team_stats['Away_Drawn']
team_stats['Total_Lost'] = team_stats['Home_Lost'] + team_stats['Away_Lost']
team_stats['Goal_Difference'] = (team_stats['Goals_Scored_Home'] + team_stats['Goals_Scored_Away']) - (team_stats['Goals_Conceded_Home'] + team_stats['Goals_Conceded_Away'])

# Drop unnecessary columns
team_stats.drop(columns=['Home_Won', 'Home_Drawn', 'Home_Lost', 'Away_Won', 'Away_Drawn', 'Away_Lost'], inplace=True)

# Sort teams by total points and overall goal difference
sorted_teams = team_stats.sort_values(by=['Total_Won', 'Total_Drawn', 'Total_Matches_Played', 'Goal_Difference'], ascending=[False, False, False, False])

# Reset the index for better readability
sorted_teams.reset_index(drop=True, inplace=True)

# Save the sorted DataFrame to a new CSV file
sorted_teams.to_csv('team_statistics.csv', index=False)

# Optionally, display the sorted DataFrame
print(sorted_teams)
```
